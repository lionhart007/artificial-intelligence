```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

# New Antecedent/Consequent objects hold universe variables and membership
# functions
pasien = ctrl.Antecedent(np.arange(0, 401, 1), 'pasien')
fasilitas_rs = ctrl.Antecedent(np.arange(0, 1001, 1), 'fasilitas_rs')
tenaga_kerja = ctrl.Antecedent(np.arange(0, 201, 1), 'tenaga_kerja')
besar_dana = ctrl.Consequent(np.arange(0, 21, 1), 'dana')

# Auto-membership function population is possible with .automf(3, 5, or 7)

# Custom membership functions can be built interactively with a familiar,
# Pythonic API

#Input 1
pasien['sedikit'] = fuzz.trapmf(pasien.universe, [0, 0, 100, 200])
pasien['sedang'] = fuzz.trimf(pasien.universe, [100, 200, 300])
pasien['banyak'] = fuzz.trapmf(pasien.universe, [200, 300, 400, 400])

#Input 2
fasilitas_rs['kurang'] = fuzz.trapmf(fasilitas_rs.universe, [0, 0, 300, 600])
fasilitas_rs['cukup'] = fuzz.trimf(fasilitas_rs.universe, [300, 600, 900])
fasilitas_rs['lengkap'] = fuzz.trapmf(fasilitas_rs.universe, [600, 900, 1000, 1000])

#Input 3
tenaga_kerja['sedikit'] = fuzz.trapmf(tenaga_kerja.universe, [0, 0, 50, 100])
tenaga_kerja['cukup'] = fuzz.trimf(tenaga_kerja.universe, [50, 100, 150])
tenaga_kerja['banyak'] = fuzz.trapmf(tenaga_kerja.universe, [100, 150, 200, 200])

#Output 1
besar_dana['sedikit'] = fuzz.trapmf(besar_dana.universe, [0, 0, 5, 10])
besar_dana['sedang'] = fuzz.trimf(besar_dana.universe, [5, 10, 15])
besar_dana['banyak'] = fuzz.trapmf(besar_dana.universe, [10, 15, 20, 20])
```

```python
rule1  = ctrl.Rule(pasien['banyak']  & fasilitas_rs['kurang']  & tenaga_kerja['sedikit'], besar_dana['banyak'])
rule2  = ctrl.Rule(pasien['banyak']  & fasilitas_rs['kurang']  & tenaga_kerja['cukup'],   besar_dana['banyak'])
rule3  = ctrl.Rule(pasien['banyak']  & fasilitas_rs['kurang']  & tenaga_kerja['banyak'],  besar_dana['banyak'])
rule4  = ctrl.Rule(pasien['banyak']  & fasilitas_rs['cukup']   & tenaga_kerja['sedikit'], besar_dana['banyak'])
rule5  = ctrl.Rule(pasien['banyak']  & fasilitas_rs['cukup']   & tenaga_kerja['cukup'],   besar_dana['banyak'])
rule6  = ctrl.Rule(pasien['banyak']  & fasilitas_rs['cukup']   & tenaga_kerja['banyak'],  besar_dana['sedikit'])
rule7  = ctrl.Rule(pasien['banyak']  & fasilitas_rs['lengkap'] & tenaga_kerja['sedikit'], besar_dana['sedang'])
rule8  = ctrl.Rule(pasien['banyak']  & fasilitas_rs['lengkap'] & tenaga_kerja['cukup'],   besar_dana['sedang'])
rule9  = ctrl.Rule(pasien['banyak']  & fasilitas_rs['lengkap'] & tenaga_kerja['banyak'],  besar_dana['sedikit'])
rule10 = ctrl.Rule(pasien['sedang']  & fasilitas_rs['kurang']  & tenaga_kerja['sedikit'], besar_dana['banyak'])
rule11 = ctrl.Rule(pasien['sedang']  & fasilitas_rs['kurang']  & tenaga_kerja['cukup'],   besar_dana['banyak'])
rule12 = ctrl.Rule(pasien['sedang']  & fasilitas_rs['kurang']  & tenaga_kerja['banyak'],  besar_dana['sedang'])
rule13 = ctrl.Rule(pasien['sedang']  & fasilitas_rs['cukup']   & tenaga_kerja['sedikit'], besar_dana['banyak'])
rule14 = ctrl.Rule(pasien['sedang']  & fasilitas_rs['cukup']   & tenaga_kerja['cukup'],   besar_dana['banyak'])
rule15 = ctrl.Rule(pasien['sedang']  & fasilitas_rs['cukup']   & tenaga_kerja['banyak'],  besar_dana['sedikit'])
rule16 = ctrl.Rule(pasien['sedang']  & fasilitas_rs['lengkap'] & tenaga_kerja['sedikit'], besar_dana['sedang'])
rule17 = ctrl.Rule(pasien['sedang']  & fasilitas_rs['lengkap'] & tenaga_kerja['cukup'],   besar_dana['sedang'])
rule18 = ctrl.Rule(pasien['sedang']  & fasilitas_rs['lengkap'] & tenaga_kerja['banyak'],  besar_dana['sedikit'])
rule19 = ctrl.Rule(pasien['sedikit'] & fasilitas_rs['kurang']  & tenaga_kerja['sedikit'], besar_dana['sedang'])
rule20 = ctrl.Rule(pasien['sedikit'] & fasilitas_rs['kurang']  & tenaga_kerja['cukup'],   besar_dana['sedang'])
rule21 = ctrl.Rule(pasien['sedikit'] & fasilitas_rs['kurang']  & tenaga_kerja['banyak'],  besar_dana['sedang'])
rule22 = ctrl.Rule(pasien['sedikit'] & fasilitas_rs['cukup']   & tenaga_kerja['sedikit'], besar_dana['sedang'])
rule23 = ctrl.Rule(pasien['sedikit'] & fasilitas_rs['cukup']   & tenaga_kerja['cukup'],   besar_dana['sedang'])
rule24 = ctrl.Rule(pasien['sedikit'] & fasilitas_rs['cukup']   & tenaga_kerja['banyak'],  besar_dana['sedikit'])
rule25 = ctrl.Rule(pasien['sedikit'] & fasilitas_rs['lengkap'] & tenaga_kerja['sedikit'], besar_dana['sedikit'])
rule26 = ctrl.Rule(pasien['sedikit'] & fasilitas_rs['lengkap'] & tenaga_kerja['cukup'],   besar_dana['sedikit'])
rule27 = ctrl.Rule(pasien['sedikit'] & fasilitas_rs['lengkap'] & tenaga_kerja['banyak'],  besar_dana['sedikit'])
#rule1.view()
```

```python
besardana_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8, rule9, rule10, rule11, rule12, rule13, rule14, rule15, rule16, rule17, rule18, rule19, rule20, rule21, rule22, rule23, rule24, rule25, rule26, rule27])
besardana = ctrl.ControlSystemSimulation(besardana_ctrl)

# Pass inputs to the ControlSystem using Antecedent labels with Pythonic API
# Note: if you like passing many inputs all at once, use .inputs(dict_of_data)
besardana.input['pasien'] = 250
besardana.input['fasilitas_rs'] = 350
besardana.input['tenaga_kerja'] = 125

# Crunch the numbers
besardana.compute()

print (besardana.output['dana'])
besar_dana.view(sim=besardana)
```
